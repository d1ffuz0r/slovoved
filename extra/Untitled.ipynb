```python
cv = CountVectorizer(lowercase=True, ngram_range=(1, 1))
```

```python
positive_df = pd.DataFrame(['рязань'], columns=['word'])
positive_df['sentiment'] = 0

negative_df = pd.DataFrame([w for w, _ in words], columns=['word'])
negative_df['sentiment'] = 1

slovoved_df = pd.concat([positive_df, negative_df])

slovoved_vector = cv.fit_transform(slovoved_df['word'])
slovoved_vector
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    slovoved_vector,
    slovoved_df['sentiment'],
    test_size=0.3,
    random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Generation Using Multinomial Naive Bayes
clf = RandomForestClassifier().fit(X_train, Y_train)
predicted = clf.predict(X_test)
print(predicted)
print("MultinomialNB Accuracy:", metrics.accuracy_score(Y_test, predicted))
```

```python
article = """Почему началась драка?

Еще до поединка бойцы не скрывали весьма напряженных отношений. Конфликт бойцов развивался еще с 2016 года, когда из-за разногласий в ходе турнира Нурмагомедов назвал Макгрегора «цыпленком».
В апреле 2018 года произошёл арест Макгрегор с друзьями напал на стоявший на парковке в Бруклине автобус, в котором находились несколько бойцов UFC, в том числе и Нурмагомедов. Причиной стала ссора Нурмагомедова еще с одним россиянином, Артемом Лобовым, другом и спарринг-партнером Макгрегора. В результате нападения несколько спортсменов получили серьезные травмы из-за разбитого стекла. После инцидента ирландца объявила в розыск полиция. В ходе расследования тот признал свою вину и отделался пятью днями общественных работ и обязательным посещением курсов самоконтроля."""
import re
repl = re.compile(r'[^А-я0-9\- ]', re.I)
prepared_article = word_tokenize(
    re.sub(repl, '', article.lower())
)
newwo = [w.lower() for w in prepared_article]

sql = """"""
for index, w in enumerate(newwo):
    if index > 0:
        sql += 'UNION\n'
    sql += """SELECT word, replacement, similarity(word, '{0}') AS sml
  FROM words
  WHERE word % '{0}'\n""".format(w)
#print(sql)
conn = psycopg2.connect('postgresql://cyberkolhoz@localhost:5432/slovoved')
with conn.cursor() as cursor:
    cursor.execute('select set_limit(0.6);')
    cursor.execute(sql)
    print(cursor.fetchall())
conn.close()
```
